# Problem 1

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import TfidfTransformer, CountVectorizer
from nltk.corpus import stopwords
import re

comp_tech_subclasses = ['comp.graphics', 
                        'comp.os.ms-windows.misc', 
                        'comp.sys.ibm.pc.hardware', 
                        'comp.sys.mac.hardware']
                        
rec_act_subclasses = ['rec.autos', 
                      'rec.motorcycles', 
                      'rec.sport.baseball', 
                      'rec.sport.hockey']
  
dataset = fetch_20newsgroups(subset='all',
                             categories=comp_tech_subclasses+rec_act_subclasses,
                             shuffle=True,
                             random_state=42,
                             remove=('headers', 'footers', 'quotes'))

stop_words = text.ENGLISH_STOP_WORDS

vect_all = CountVectorizer(analyzer = 'word', stop_words = stop_words, min_df = 3)
counts_all = vect_all.fit_transform(dataset.data)
TFxIDF_transformer = TfidfTransformer()
X = TFxIDF_transformer.fit_transform(counts_all)

print("The dimension of TF-IDF Vector with min_df=3 is",X.shape)
```

# problem 2

```python
new_target = (dataset.target[:]<4)*1
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics

km = KMeans(n_clusters=2, init='k-means++', max_iter=500, n_init=5)
km.fit(X)

print ('Homogeneity:', metrics.homogeneity_score(new_target, km.labels_))
print ('Completeness:', metrics.completeness_score(new_target, km.labels_))
print ('V-measure:', metrics.v_measure_score(new_target, km.labels_))
print ('Adjusted Rand Score:', metrics.adjusted_rand_score(new_target, km.labels_))
print ('Adjusted Mutual Info Score:', metrics.adjusted_mutual_info_score(new_target, km.labels_))
print ('The contingency matrix:\n',metrics.confusion_matrix(new_target, km.labels_),'\n'*2)
```

```python
def plot_clusters(actual_labels, clustered_labels, X_2d):
    color = ["r", "b"]
    mark = ["o", "+"]
    for i in range(len(new_target)):
        plt.scatter(X_2d[i, 0], X_2d[i, 1], s=12, marker=mark[actual_labels[i]], color=color[clustered_labels[i]], alpha=0.5)
    plt.show()
```

## SVD

```python
svd = TruncatedSVD(n_components=10, n_iter=13, random_state=42)
lsa = make_pipeline(svd)
X_reduced = lsa.fit_transform(X)

km = KMeans(n_clusters=2, init='k-means++', max_iter=500, n_init=5)
km.fit(X_reduced)
clustered_labels = km.labels_

X_2d = svd.fit_transform(X)

plot_clusters(new_target, clustered_labels, X_2d)

print ('The contingency matrix:\n',metrics.confusion_matrix(new_target, km.labels_),'\n'*2)
```

```python
comp_tech_subclasses = ['comp.graphics',
                        'comp.os.ms-windows.misc',
                        'comp.sys.ibm.pc.hardware',
                        'comp.sys.mac.hardware',
                        'comp.windows.x']

rec_act_subclasses = ['rec.autos',
                      'rec.motorcycles',
                      'rec.sport.baseball',
                      'rec.sport.hockey']

science_subclass = ['sci.crypt', 'sci.electronics', 'sci.med', 'sci.space']

miscellaneous_subclass = ['misc.forsale']

politics_subclass = ['talk.politics.misc', 'talk.politics.guns', 'talk.politics.mideast']

religion_subclass = ['talk.religion.misc', 'alt.atheism', 'soc.religion.christian']

dataset = fetch_20newsgroups(subset='all',
                             categories=comp_tech_subclasses + rec_act_subclasses + science_subclass +
                                        miscellaneous_subclass + politics_subclass + religion_subclass,
                             shuffle=True,
                             random_state=42,
                             remove=('headers', 'footers', 'quotes'))

stop_words = text.ENGLISH_STOP_WORDS

vect_20 = CountVectorizer(analyzer = 'word', stop_words = stop_words, min_df = 3)
counts_20 = vect_20.fit_transform(dataset.data)
TFxIDF_transformer = TfidfTransformer()
X_20 = TFxIDF_transformer.fit_transform(counts_20)

print("The dimension of TF-IDF Vector with min_df=3 is",X_20.shape)
```

```python
## SVD 
homo = []
complet = []
vmeasure = []
rands = []
mutual = []
valuee = [1,2,3,5,10,20,50,100]

for i in valuee:
    
    print('r =',i)
    
    svd = TruncatedSVD(n_components = i, n_iter = 13,random_state = 42)
    lsa = make_pipeline(svd)    
    X_reduced = lsa.fit_transform(X_20)    
    km = KMeans(n_clusters=20, init='k-means++', max_iter=500, n_init=5)
    km.fit(X_reduced)
    
    homo.append(metrics.homogeneity_score(dataset.target, km.labels_))
    complet.append(metrics.completeness_score(dataset.target, km.labels_))
    vmeasure.append(metrics.v_measure_score(dataset.target, km.labels_))
    rands.append(metrics.adjusted_rand_score(dataset.target, km.labels_))
    mutual.append(metrics.adjusted_mutual_info_score(dataset.target, km.labels_))
    
    print ('Homogeneity:', metrics.homogeneity_score(dataset.target, km.labels_))
    print ('Completeness:', metrics.completeness_score(dataset.target, km.labels_))
    print ('V-measure:', metrics.v_measure_score(dataset.target, km.labels_))
    print ('Adjusted Rand Score:', metrics.adjusted_rand_score(dataset.target, km.labels_))
    print ('Adjusted Mutual Info Score:', metrics.adjusted_mutual_info_score(dataset.target, km.labels_))
    print ('The contingency matrix:\n',metrics.confusion_matrix(dataset.target, km.labels_),'\n'*2)
    print('----------------------------------')
    
plt.plot(valuee, homo,valuee, complet,valuee, vmeasure,valuee, rands,valuee, mutual)
plt.xlabel('r', fontsize = 20)
plt.legend(('Homogeneity', 'Completeness', 'V-measure', 'Adjusted Rand Score','Adjusted Mutual Info Score'))
plt.show()

```
